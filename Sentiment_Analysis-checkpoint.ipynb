## Step 2. Vader Sentiment 

We'll utilize NLTK's SentimentIntensityAnalyzer to obtain the negative, neutral, and positive scores of the text.

=>This employs a "bag of words" technique:
   1. Stop words are eliminated,
   2. Individual words are assigned scores, and then aggregated into a total score.

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm 
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores(text_ex)
```

```python
# Run the polarity score on the entire dataset
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    text = row['Text']
    myid = row['Id']
    res[myid] = sia.polarity_scores(text)
```

```python
vaders = pd.DataFrame(res).T
vaders = vaders.reset_index().rename(columns={'index': 'Id'})
vaders = vaders.merge(df, how='left')
```

```python
vaders.head()
```

### Plotting VADER Result

## Step 3. Roberta Pretrained Model
1. Employ a model trained on a vast corpus of data.
2. The Transformer model considers not only individual words but also the contextual relationships with other words.

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
def polarity_scores_roberta(example):
    encoded_text = tokenizer(example, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict
```

```python
res = {}
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        text = row['Text']
        myid = row['Id']
        vader_result = sia.polarity_scores(text)
        vader_result_rename = {}
        for key, value in vader_result.items():
            vader_result_rename[f"vader_{key}"] = value
        roberta_result = polarity_scores_roberta(text)
        both = {**vader_result_rename, **roberta_result}
        res[myid] = both
    except RuntimeError:
        print(f'Broke for id {myid}')
```

```python
results_df = pd.DataFrame(res).T
results_df = results_df.reset_index().rename(columns={'index': 'Id'})
results_df = results_df.merge(df, how='left')
```

### Step 4. Combine and comapre 

```python
sns.pairplot(data=results_df,
             vars=['vader_neg', 'vader_neu', 'vader_pos',
                  'roberta_neg', 'roberta_neu', 'roberta_pos'],
            hue='Score',
            palette='tab10')
plt.show()
```

```python
# Count the occurrences of each sentiment category for Roberta analysis
roberta_pos_count = results_df[results_df['roberta_pos'] > 0.5].shape[0]
roberta_neg_count = results_df[results_df['roberta_neg'] > 0.5].shape[0]
roberta_neu_count = results_df[results_df['roberta_neu'] > 0.5].shape[0]
```

```python
y1 = np.array([roberta_pos_count/results_df.shape[0], roberta_neg_count/results_df.shape[0], roberta_neu_count/results_df.shape[0]])
mylabels = ["positive_reviews", "negative_reviews", "neutral_reviews"]
mycolors = ["green", "orange", "red"]
plt.pie(y1, labels=mylabels, colors=mycolors, autopct='%1.1f%%')
plt.show()
```

```python
# Count the occurrences of each sentiment category for VADER analysis
pos_count = results_df[results_df['vader_compound'] > 0].shape[0]
neg_count = results_df[results_df['vader_compound'] < 0].shape[0]
neu_count = results_df[results_df['vader_compound'] == 0].shape[0]
```

```python
y = np.array([pos_count/(pos_count+neg_count+neu_count), neg_count/(pos_count+neg_count+neu_count), neu_count/(pos_count+neg_count+neu_count)])
mylabels=["positive_reviews", "negative_reviews", "neutral_reviews"]
mycolors=["green", "orange", "red"]
plt.pie(y, labels=mylabels, colors=mycolors, autopct='%1.1f%%')
plt.show()
```
